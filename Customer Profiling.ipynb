**We will Procced with model building now and train a KNN classifier model**

Separate the dependent variable from the features

```python
# Define fetures in X and Output in Y

X = df1.iloc[:, :-1].values
y = df1.iloc[:, -1].values
```

**We will convert features and label columns into binary Encoding**

```python
# Feature encoding (Independednt variable)#Onehot encoder isused only for features

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print (X)
```

```python
# Label encoding (dependent vriable) is used only for output

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

We will split the 'X' and 'y' data into **Test and Train**

```python
#Splitting df1 into Train and Test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

Using standard scaler we will convert all the values into a standard scale.

```python
#convert all numerical values to a standdard scale
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train= sc.fit_transform(x_train[:, 13:])
x_test= sc.transform(x_test[:,13:])
```

**Model Building**

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
acc_list = []
err_list = []
for i in range(1,25):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    acc = accuracy_score(y_test,y_pred)
    err = 1-acc
    acc_list.append(acc)
    err_list.append(err)
    print("For k = {}, accuracy = {}".format(i,acc))
    print("For k = {}, error = {}".format(i,err))
    
```

```python
#Plot line chart for error rate
plt.plot(list(range(1,25)),err_list,c= 'r')
plt.title('Error rate v/s K')
plt.xlabel('K')
plt.ylabel('Error rate')
plt.show()
```

From the above plot we have concluded that K= 18 gives us the highest accuracy

We train the Knn claissifier model with K=18

```python
#Model building
model = KNeighborsClassifier(n_neighbors=18)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

We can evaluate the confusion matrix

```python
#Model Evaluation
confusion_matrix(y_test,y_pred)
```

```python
#accuracy check
a=accuracy_score(y_test,y_pred)
print(a*100,'%')
```

From this model we have got accuracy of 86.71%
